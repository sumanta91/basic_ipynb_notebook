```python
import numpy as np
import pylab as plb
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math

tau=1
tpi=0.3

#t and tpi all are in \mu sec

t=np.arange(0., 1*np.pi, 0.03)
z=np.exp(-(t*1e-6/(tau*1e-6))**2/2)
def p(x): return np.cos((np.pi/(tpi*1e-6))*1e-6*x)
A=1+z*p(t)

plt.figure(figsize=(10,6))
plt.plot(t,0.5*A,'o-',label='data')
#plt.plot(x,gaus(x*popt),'ro:',label='fit')
plt.xlabel('time')
plt.show()

```
